#### **Select**

```python
filtered_movies = df.loc[
    (df['Year'].between(2012, 2014)) & 
    (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.65))
]

final_movie_list = filtered_movies[
    filtered_movies['Director'].isin(
        filtered_movies['Director'].value_counts().loc[lambda x: x >= 2].index
    )
].copy()

final_movie_list
```

#### **Groupby**

```python
final_movie_list['Num of movies']   = final_movie_list.groupby('Director')['Director'].transform('count')
final_movie_list['Rating category'] = final_movie_list.groupby('Director')['Rating'].transform('mean').apply(lambda x: 'Good' if x >= 8 else 'Bad')
final_movie_list
```

```python
groupby_num_movies_directors = final_movie_list.groupby(['Rating category', 'Num of movies', 'Director']).agg({
    'Revenue (Millions)': lambda x: round(x.mean(), 2),
    'Rating': ['max', 'min']
})

sorted_groupby = groupby_num_movies_directors.sort_values(
    by=['Rating category', 'Num of movies', ('Revenue (Millions)', '<lambda>')],
    ascending=[1, 1, 0] 
)

sorted_groupby
```
