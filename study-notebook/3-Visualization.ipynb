```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/3-iris.csv')
df.head()
```

## **1. Scatter Plot**

- **Purpose**: Visualizes the relationship between two continuous variables, helping to identify correlations.
- **Common Use Case**: Frequently used in regression analysis and to visualize feature relationships in Machine Learning.
- **Data Type**: Two continuous variables.
- **Key Parameters**:
  - `x`: The variable to plot on the x-axis (continuous).
  - `y`: The variable to plot on the y-axis (continuous).
  - `data`: The dataset that contains the variables.
  - `hue`: (Optional) Categorical variable to color the points, allowing the visualization of groupings.

```python
_, ax = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='species', 
                ax=ax[0], palette='Set2', s=100)

ax[0].set_title('Sepal length vs. Sepal width', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Sepal Length (cm)', fontsize=12)
ax[0].set_ylabel('Sepal Width (cm)', fontsize=12)
ax[0].legend(title='Species', loc='upper right')

sns.scatterplot(x='petal_length', y='petal_width', data=df, hue='species', 
                ax=ax[1], palette='Set2', s=100)

ax[1].set_title('Petal length vs. Petal width', fontsize=14, fontweight='bold')
ax[1].set_xlabel('Petal Length (cm)', fontsize=12)
ax[1].set_ylabel('Petal Width (cm)', fontsize=12)
ax[1].legend(title='Species', loc='upper left')

plt.tight_layout()
plt.show()
```

## 2. Line Plot

- **Purpose**: Displays trends in data over a continuous axis, making it ideal for tracking changes over time.
- **Common Use Case**: Commonly used to track performance metrics such as loss or accuracy during model training.
- **Data Type**: Continuous data, or time-series data if analyzing trends over time.
- **Key Parameters**:
  - `x`: The variable to plot on the x-axis (continuous or time).
  - `y`: The variable to plot on the y-axis (continuous).
  - `data`: The dataset that contains the variables.
  - `hue`: (Optional) Categorical variable to differentiate lines based on a category.

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=df, hue='species', 
             palette='pastel', linewidth=2.5)

plt.title('Sepal Length vs Sepal Width (by Species)', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.grid(True)
plt.legend(title='Species')
plt.show()
```

## 3. Bar Plot

- **Purpose**: Compares different categories by displaying the values of a continuous variable.
- **Common Use Case**: Often used in classification tasks to compare model performance metrics across different classes.
- **Data Type**: Categorical data on x-axis and continuous data on y-axis.
- **Key Parameters**:
  - `x`: Categorical variable to plot on the x-axis.
  - `y`: Continuous variable to plot on the y-axis (mean or sum by default).
  - `data`: The dataset that contains the variables.
  - `hue`: (Optional) Categorical variable to show grouped bars.

```python
sns.barplot(x='species', y='sepal_length', data=df, hue='species', 
            palette='pastel', legend=False)

plt.title('Average Sepal Length per Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.show()
```

## 4. Box Plot

- **Purpose**: Summarizes data distribution based on quartiles, providing insight into the spread and outliers.
- **Common Use Case**: Frequently used to detect outliers and compare feature distributions across different classes.
- **Data Type**: Continuous variable (y) and categorical variable (x).
- **Key Parameters**:
  - `x`: Categorical variable to group the continuous data.
  - `y`: Continuous variable to plot.
  - `data`: The dataset that contains the variables.
  - `hue`: (Optional) Categorical variable to show additional grouping.

```python
sns.boxplot(x='species', y='sepal_width', data=df, hue='species', 
            palette='pastel', legend=False)

plt.title('Distribution of Sepal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.show()
```

## 5. Histogram

- **Purpose**: Displays the frequency distribution of a continuous variable.
- **Common Use Case**: Typically used in exploratory data analysis (EDA) to examine the distribution of features and detect skewness.
- **Data Type**: Single continuous variable.
- **Key Parameters**:
  - `x`: The variable to plot.
  - `data`: The dataset that contains the variable.
  - `bins`: (Optional) Number of bins to divide the data. More bins provide higher granularity, fewer bins give a more generalized view.
  - `kde`: (Optional) Boolean to add Kernel Density Estimate curve, which smooths out the distribution.

```python
sns.histplot(x='sepal_width', data=df, bins=20, kde=True, 
             color='skyblue')

plt.title('Distribution of Sepal Width', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Width (cm)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

## 6. Heatmap

- **Purpose**: Visualizes relationships between variables in matrix form, making it useful for analyzing correlations.
- **Common Use Case**: Commonly used to visualize correlation matrices in feature selection for regression tasks.
- **Data Type**: Matrix of numeric data (e.g., correlation matrix).
- **Key Parameters**:
  - `data`: The matrix to visualize.
  - `annot`: (Optional) Boolean to display values inside each cell.
  - `cmap`: (Optional) Color map that defines the color scale of the heatmap.

```python
sns.heatmap(data=df.corr(), annot=True, cmap='Blues', linewidths=0.5)

plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

## 7. Pair Plot

- **Purpose**: Displays pairwise relationships between variables along with their distributions.
- **Common Use Case**: Used in exploratory data analysis (EDA) to visualize interactions between multiple features in datasets.
- **Data Type**: Continuous data (with optional categorical grouping).
- **Key Parameters**:
  - `data`: The dataset containing the variables.
  - `hue`: (Optional) Categorical variable to color the points by class or group.

```python
sns.pairplot(data=df, hue='species', palette='husl', corner=False)

plt.suptitle('Pairplot of Iris Dataset', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```

## 8. Violin Plot (Biểu đồ violin)

- **Purpose**: Combines features of a box plot and kernel density plot to show the distribution of data.
- **Common Use Case**: Frequently used to visualize the distribution of features across different classes in classification tasks.
- **Data Type**: Continuous variable (y) grouped by a categorical variable (x).
- **Key Parameters**:
  - `x`: Categorical variable.
  - `y`: Continuous variable to plot.
  - `data`: The dataset that contains the variables.
  - `hue`: (Optional) Categorical variable for further grouping.

```python
sns.violinplot(x='species', y='sepal_width', data=df, hue='species', 
               palette='pastel', split=False)

plt.title('Violin Plot of Sepal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.show()
```

## 9. Count Plot

- **Purpose**: Visualizes the count of occurrences of categories in a dataset.
- **Common Use Case**: Often used in classification problems to check for class imbalances or analyze categorical distributions.
- **Data Type**: Categorical data.
- **Key Parameters**:
  - `x`: Categorical variable to count occurrences for.
  - `data`: The dataset containing the categorical variable.
  - `hue`: (Optional) Categorical variable to group the counts by another category.

```python
sns.countplot(x='petal_width', data=df, hue='species', palette='pastel')

plt.title('Count of Petal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Petal Width (cm)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Species')
plt.show()
```
